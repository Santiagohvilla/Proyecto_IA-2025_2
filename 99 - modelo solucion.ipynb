```python
d1['E_PRGM_DEPARTAMENTO'] = d1['E_PRGM_DEPARTAMENTO'].str.upper() # A mayúsculas para coincidir con la lista
d1['E_PRGM_DEPARTAMENTO'] = d1['E_PRGM_DEPARTAMENTO'].str.replace('[^A-Z\s]', '', regex=True)
d1['E_PRGM_DEPARTAMENTO'] = d1['E_PRGM_DEPARTAMENTO'].str.replace('\s+', ' ', regex=True).str.strip()

REGIONES = {
    'ANDINA': [
        'ANTIOQUIA', 'BOYACA', 'BOGOT', 'BOGOTA','CALDAS', 'CUNDINAMARCA', 'HUILA', 'NORTE SANTANDER',
        'QUINDIO', 'RISARALDA', 'SANTANDER', 'TOLIMA', 'NARINO','NARIO'
    ],
    'CARIBE': [
        'ATLANTICO', 'BOLIVAR', 'CESAR', 'CORDOBA', 'LA GUAJIRA', 'MAGDALENA',
        'SUCRE', 'SAN ANDRES'
    ],
    'PACIFICA': [
        'CHOCO', 'VALLE', 'CAUCA'
    ],
    'ORINOQUIA': [
        'ARAUCA', 'CASANARE', 'META', 'VICHADA'
    ],
    'AMAZONIA': [
        'AMAZONAS', 'CAQUETA', 'GUAINIA', 'GUAVIARE', 'PUTUMAYO', 'VAUPES'
    ]}
for region, departamentos in REGIONES.items():
    # Asigna la clave del diccionario (ej., 'ANDINA') a todas las filas
    # donde el valor de 'Departamento' está DENTRO de la lista (ej., la lista de departamentos Andinos)
    d1.loc[d1['E_PRGM_DEPARTAMENTO'].isin(departamentos), 'E_PRGM_DEPARTAMENTO'] = region

d1=pd.get_dummies(d1, columns=['E_PRGM_DEPARTAMENTO'], prefix='E_PRGM')

bool_cols = d1.select_dtypes(include=['bool']).columns
d1[bool_cols] = d1[bool_cols].astype(int)
```

```python
def replace_category_for_numbers(df, column_name, mapping_dict):
    df[column_name] = df[column_name].map(mapping_dict)
    return df

mapping_dict = {
    'bajo': 1,
    'medio-bajo': 2,
    'medio-alto': 3,
    'alto': 4}
d2=d1.copy()
d2 = replace_category_for_numbers(d2, 'RENDIMIENTO_GLOBAL', mapping_dict)

mapping_estrato = {
    "estrato1": 1,
    "estrato2": 2,
    "estrato3": 3,
    "estrato4": 4,
    "estrato5": 5,
    "estrato6": 6,
    "sinestrato": 0
}
d2['F_ESTRATOVIVIENDA'] = (
    d2['F_ESTRATOVIVIENDA']
    .astype(str)
    .str.strip()
    .str.replace(" ", "", regex=False)   # Eliminar espacios
    .str.lower()
)

d2 = replace_category_for_numbers(d2, "F_ESTRATOVIVIENDA", mapping_estrato)
#Se llenan casillas vacias con la moda de estrato
d2['F_ESTRATOVIVIENDA'] = d2['F_ESTRATOVIVIENDA'].fillna(d2['F_ESTRATOVIVIENDA'].mode()[0])

d3=d2.copy()
columnas_a_limpiar = ['F_EDUCACIONMADRE', 'F_EDUCACIONPADRE']

for col in columnas_a_limpiar:
    d3[col] = (
        d3[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )

mapping_edu_padres = {
    "ninguno": 0,
    "no aplica": 0,
    "no sabe": 0,
    "primaria incompleta": 1,
    "primaria completa": 2,
    "secundaria (bachillerato) incompleta": 3,
    "secundaria (bachillerato) completa": 4,
    "técnica o tecnológica incompleta": 5,
    "tecnica o tecnologica incompleta": 5,
    "técnica o tecnológica completa": 6,
    "tecnica o tecnologica completa": 6,
    "educación profesional incompleta": 7,
    "educacion profesional incompleta": 7,
    "educación profesional completa": 8,
    "educacion profesional completa": 8,
    "postgrado": 9
}


d3 = replace_category_for_numbers(d3, "F_EDUCACIONMADRE", mapping_edu_padres)
d3 = replace_category_for_numbers(d3, "F_EDUCACIONPADRE", mapping_edu_padres)
#Se llenan casillas vacias con 0
d3['F_EDUCACIONMADRE'] = d3['F_EDUCACIONMADRE'].fillna(0)
d3['F_EDUCACIONPADRE'] = d3['F_EDUCACIONPADRE'].fillna(0)

d4=d3.copy()
def clean_text_column(df, column_name):
    df[column_name] = (
        df[column_name]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.replace('á', 'a')
        .str.replace('é', 'e')
        .str.replace('í', 'i')
        .str.replace('ó', 'o')
        .str.replace('ú', 'u')
        .str.replace(r'[^a-z0-9\s]', '', regex=True)  # ✅ cadena cruda para regex
    )
    df[column_name] = df[column_name].replace('nan', np.nan)
    return df

mapping_matri = {
    'no pago matricula': 0,
    'menos de 500 mil': 1,
    'entre 500 mil y menos de 1 millon': 2,
    'entre 1 millon y menos de 25 millones': 3,
    'entre 25 millones y menos de 4 millones': 4,
    'entre 4 millones y menos de 55 millones': 5,
    'entre 55 millones y menos de 7 millones': 6,
    'mas de 7 millones': 7
}

d4 = clean_text_column(d4, 'E_VALORMATRICULAUNIVERSIDAD')
d4 = replace_category_for_numbers(d4, 'E_VALORMATRICULAUNIVERSIDAD', mapping_matri)
#llenar casillas vacias con valor promedio
d4['E_VALORMATRICULAUNIVERSIDAD'] = d4['E_VALORMATRICULAUNIVERSIDAD'].fillna(round(d4['E_VALORMATRICULAUNIVERSIDAD'].mean()))

def clean_text_column(df, column):
    df[column] = (
        df[column]
        .astype(str)
        .str.lower()
        .str.normalize('NFKD')
        .str.encode('ascii', errors='ignore')
        .str.decode('utf-8')
        .str.replace(r'[^a-z0-9\s]', '', regex=True)
        .str.replace(r'\s+', '', regex=True)
    )
    return df

mapping_work = {
    '0': 1,
    'menosde10horas': 2,
    'entre11y20horas': 3,
    'entre21y30horas': 4,
    'masde30horas': 5,
}

d5=d4.copy()
d5 = clean_text_column(d5, 'E_HORASSEMANATRABAJA')
d5 = replace_category_for_numbers(d5, 'E_HORASSEMANATRABAJA', mapping_work)
#Se rellenan casillas vacias con el promedio
d5['E_HORASSEMANATRABAJA'] = d5['E_HORASSEMANATRABAJA'].fillna(round(d5['E_HORASSEMANATRABAJA'].mean()))

d6 = d5.copy()

mapping_binario = {
    'si': 1,
    'no': 0
}

cols_binarias = ['F_TIENELAVADORA','F_TIENEINTERNET','F_TIENEAUTOMOVIL', 'E_PAGOMATRICULAPROPIO', 'F_TIENECOMPUTADOR']

for col in cols_binarias:
    d6 = clean_text_column(d6, col)
    d6 = replace_category_for_numbers(d6, col, mapping_binario)

#llenar casillas vacias de  con moda
for col in cols_binarias:
    moda = d6[col].mode(dropna=True)[0]
    d6[col] = d6[col].fillna(moda)

mapping_privado = {
    's': 1,
    'n': 0
}

d6 = clean_text_column(d6, 'E_PRIVADO_LIBERTAD')
d6 = replace_category_for_numbers(d6, 'E_PRIVADO_LIBERTAD', mapping_privado)
#llenar casillas vacias de  con moda
moda = d6['E_PRIVADO_LIBERTAD'].mode(dropna=True)[0]
d6['E_PRIVADO_LIBERTAD'] = d6['E_PRIVADO_LIBERTAD'].fillna(moda)
```

# PROBAR MODELOS

```python

from sklearn.metrics import median_absolute_error, r2_score, mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

estimator1 = DecisionTreeRegressor(max_depth=5)
estimator2 = DecisionTreeRegressor(max_depth=10)
estimator3 = LinearRegression()
#estimator4 = RandomForestRegressor(n_estimators=10)
estimators = [estimator1, estimator2, estimator3]

test_size = 0.3 # 30% de los datos serán para validar el modelo y 70% para entrenarlo
val_size  = test_size/(1-test_size)
Xtv, Xts, ytv, yts = train_test_split(X, y, test_size=test_size)
print(Xtv.shape, Xts.shape, ytv.shape, yts.shape)
```

```python
from sklearn.model_selection import cross_validate, ShuffleSplit
#METRICA
def rel_mrae(estimator, X, y):
    preds = estimator.predict(X)
    return np.mean(np.abs(preds-y)/y)

def report_cv_score(z):
    print ("test score   %.3f (±%.4f) with %d splits"%(np.mean(z["test_score"]), np.std(z["test_score"]), len(z["test_score"])))
    print ("train score  %.3f (±%.4f) with %d splits"%(np.mean(z["train_score"]), np.std(z["train_score"]), len(z["train_score"])))

zscores = []
for estimator in estimators:
    print("--")
    z = cross_validate(estimator, Xtv, ytv, return_train_score=True, return_estimator=False,
                       scoring=rel_mrae, cv=ShuffleSplit(n_splits=10, test_size=val_size))
    report_cv_score(z)
    zscores.append(np.mean(z["test_score"]))

best = np.argmin(zscores)
print ("selecting ", best)
best_estimator = estimators[best]
print ("\nselected model")
print (best_estimator)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['rbf'] # Mantenemos el kernel que es más versátil
}

grid_search = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    scoring=rel_mrae, # Usar tu métrica personalizada
    cv=ShuffleSplit(n_splits=10, test_size=val_size),
    verbose=1, # Para ver el progreso
    n_jobs=-1 # Usar todos los núcleos del CPU
)

grid_search.fit(Xtv, ytv)

# 4. Obtener la mejor configuración
best_C_gamma = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Mejores parámetros encontrados: {best_C_gamma}")
print(f"Mejor rendimiento CV: {best_score}")

# El modelo óptimo para usar es:
best_svc_model = grid_search.best_estimator_
```

```python
def replace_category_for_numbers(df, column_name, mapping_dict):
    df[column_name] = df[column_name].map(mapping_dict)
    return df
d2=d7.copy()

mapping_estrato = {
    "estrato1": 1,
    "estrato2": 2,
    "estrato3": 3,
    "estrato4": 4,
    "estrato5": 5,
    "estrato6": 6,
    "sinestrato": 0
}
d2['F_ESTRATOVIVIENDA'] = (
    d2['F_ESTRATOVIVIENDA']
    .astype(str)
    .str.strip()
    .str.replace(" ", "", regex=False)   # Eliminar espacios
    .str.lower()
)

d2 = replace_category_for_numbers(d2, "F_ESTRATOVIVIENDA", mapping_estrato)
#Se llenan casillas vacias con la moda de estrato, manejando el caso de moda vacía
mode_estrato = d2['F_ESTRATOVIVIENDA'].mode()
if not mode_estrato.empty:
    d2['F_ESTRATOVIVIENDA'] = d2['F_ESTRATOVIVIENDA'].fillna(mode_estrato[0])
else:
    d2['F_ESTRATOVIVIENDA'] = d2['F_ESTRATOVIVIENDA'].fillna(0) # Default to 'sinestrato'

d3=d2.copy()
columnas_a_limpiar = ['F_EDUCACIONMADRE', 'F_EDUCACIONPADRE']

for col in columnas_a_limpiar:
    d3[col] = (
        d3[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )

mapping_edu_padres = {
    "ninguno": 0,
    "no aplica": 0,
    "no sabe": 0,
    "primaria incompleta": 1,
    "primaria completa": 2,
    "secundaria (bachillerato) incompleta": 3,
    "secundaria (bachillerato) completa": 4,
    "técnica o tecnológica incompleta": 5,
    "tecnica o tecnologica incompleta": 5,
    "técnica o tecnológica completa": 6,
    "tecnica o tecnologica completa": 6,
    "educación profesional incompleta": 7,
    "educacion profesional incompleta": 7,
    "educación profesional completa": 8,
    "educacion profesional completa": 8,
    "postgrado": 9
}


d3 = replace_category_for_numbers(d3, "F_EDUCACIONMADRE", mapping_edu_padres)
d3 = replace_category_for_numbers(d3, "F_EDUCACIONPADRE", mapping_edu_padres)
#Se llenan casillas vacias con 0
d3['F_EDUCACIONMADRE'] = d3['F_EDUCACIONMADRE'].fillna(0)
d3['F_EDUCACIONPADRE'] = d3['F_EDUCACIONPADRE'].fillna(0)

d4=d3.copy()
def clean_text_column(df, column_name):
    df[column_name] = (
        df[column_name]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.replace('á', 'a')
        .str.replace('é', 'e')
        .str.replace('í', 'i')
        .str.replace('ó', 'o')
        .str.replace('ú', 'u')
        .str.replace(r'[^a-z0-9\s]', '', regex=True)  # ✅ cadena cruda para regex
    )
    df[column_name] = df[column_name].replace('nan', np.nan)
    return df

mapping_matri = {
    'no pago matricula': 0,
    'menos de 500 mil': 1,
    'entre 500 mil y menos de 1 millon': 2,
    'entre 1 millon y menos de 25 millones': 3,
    'entre 25 millones y menos de 4 millones': 4,
    'entre 4 millones y menos de 55 millones': 5,
    'entre 55 millones y menos de 7 millones': 6,
    'mas de 7 millones': 7
}

d4 = clean_text_column(d4, 'E_VALORMATRICULAUNIVERSIDAD')
d4 = replace_category_for_numbers(d4, 'E_VALORMATRICULAUNIVERSIDAD', mapping_matri)
#llenar casillas vacias con valor promedio
d4['E_VALORMATRICULAUNIVERSIDAD'] = d4['E_VALORMATRICULAUNIVERSIDAD'].fillna(round(d4['E_VALORMATRICULAUNIVERSIDAD'].mean()))

def clean_text_column(df, column):
    df[column] = (
        df[column]
        .astype(str)
        .str.lower()
        .str.normalize('NFKD')
        .str.encode('ascii', errors='ignore')
        .str.decode('utf-8')
        .str.replace(r'[^a-z0-9\s]', '', regex=True)
        .str.replace(r'\s+', '', regex=True)
    )
    return df

mapping_work = {
    '0': 1,
    'menosde10horas': 2,
    'entre11y20horas': 3,
    'entre21y30horas': 4,
    'masde30horas': 5,
}

d5=d4.copy()
d5 = clean_text_column(d5, 'E_HORASSEMANATRABAJA')
d5 = replace_category_for_numbers(d5, 'E_HORASSEMANATRABAJA', mapping_work)
#Se rellenan casillas vacias con el promedio
d5['E_HORASSEMANATRABAJA'] = d5['E_HORASSEMANATRABAJA'].fillna(round(d5['E_HORASSEMANATRABAJA'].mean()))

d6 = d5.copy()

mapping_binario = {
    'si': 1,
    'no': 0
}

cols_binarias = ['F_TIENELAVADORA','F_TIENEINTERNET','F_TIENEAUTOMOVIL', 'E_PAGOMATRICULAPROPIO', 'F_TIENECOMPUTADOR']

for col in cols_binarias:
    d6 = clean_text_column(d6, col)
    d6 = replace_category_for_numbers(d6, col, mapping_binario)

#llenar casillas vacias de  con moda
for col in cols_binarias:
    moda = d6[col].mode(dropna=True)
    if not moda.empty:
        d6[col] = d6[col].fillna(moda[0])
    else:
        d6[col] = d6[col].fillna(0) # Default to 0 if mode is empty

mapping_privado = {
    's': 1,
    'n': 0
}

d6 = clean_text_column(d6, 'E_PRIVADO_LIBERTAD')
d6 = replace_category_for_numbers(d6, 'E_PRIVADO_LIBERTAD', mapping_privado)
#llenar casillas vacias de  con moda
moda = d6['E_PRIVADO_LIBERTAD'].mode(dropna=True)
if not moda.empty:
    d6['E_PRIVADO_LIBERTAD'] = d6['E_PRIVADO_LIBERTAD'].fillna(moda[0])
else:
    d6['E_PRIVADO_LIBERTAD'] = d6['E_PRIVADO_LIBERTAD'].fillna(0) # Default to 0 if mode is empty
```
