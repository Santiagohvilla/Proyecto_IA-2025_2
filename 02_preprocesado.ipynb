```python
d2['E_PRGM_DEPARTAMENTO'] = d2['E_PRGM_DEPARTAMENTO'].str.upper() # A mayúsculas para coincidir con la lista
d2['E_PRGM_DEPARTAMENTO'] = d2['E_PRGM_DEPARTAMENTO'].str.replace('[^A-Z\s]', '', regex=True)
d2['E_PRGM_DEPARTAMENTO'] = d2['E_PRGM_DEPARTAMENTO'].str.replace('\s+', ' ', regex=True).str.strip()

REGIONES = {
    'ANDINA': [
        'ANTIOQUIA', 'BOYACA', 'BOGOT', 'BOGOTA','CALDAS', 'CUNDINAMARCA', 'HUILA', 'NORTE SANTANDER',
        'QUINDIO', 'RISARALDA', 'SANTANDER', 'TOLIMA', 'NARINO','NARIO'
    ],
    'CARIBE': [
        'ATLANTICO', 'BOLIVAR', 'CESAR', 'CORDOBA', 'LA GUAJIRA', 'MAGDALENA',
        'SUCRE', 'SAN ANDRES'
    ],
    'PACIFICA': [
        'CHOCO', 'VALLE', 'CAUCA'
    ],
    'ORINOQUIA': [
        'ARAUCA', 'CASANARE', 'META', 'VICHADA'
    ],
    'AMAZONIA': [
        'AMAZONAS', 'CAQUETA', 'GUAINIA', 'GUAVIARE', 'PUTUMAYO', 'VAUPES'
    ]}
for region, departamentos in REGIONES.items():
    # Asigna la clave del diccionario (ej., 'ANDINA') a todas las filas
    # donde el valor de 'Departamento' está DENTRO de la lista (ej., la lista de departamentos Andinos)
    d2.loc[d2['E_PRGM_DEPARTAMENTO'].isin(departamentos), 'E_PRGM_DEPARTAMENTO'] = region

#VERIFICAR
categorias = d2['E_PRGM_DEPARTAMENTO'].unique()
print(categorias)
```

```python
def clean_text_column(df, column_name):
    df[column_name] = (
        df[column_name]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.replace('á', 'a')
        .str.replace('é', 'e')
        .str.replace('í', 'i')
        .str.replace('ó', 'o')
        .str.replace('ú', 'u')
        .str.replace(r'[^a-z0-9\s]', '', regex=True)  # ✅ cadena cruda para regex
    )
    df[column_name] = df[column_name].replace('nan', np.nan)
    return df

mapping_matri = {
    'no pago matricula': 0,
    'menos de 500 mil': 1,
    'entre 500 mil y menos de 1 millon': 2,
    'entre 1 millon y menos de 25 millones': 3,
    'entre 25 millones y menos de 4 millones': 4,
    'entre 4 millones y menos de 55 millones': 5,
    'entre 55 millones y menos de 7 millones': 6,
    'mas de 7 millones': 7
}

d6 = clean_text_column(d6, 'E_VALORMATRICULAUNIVERSIDAD')
d6 = replace_category_for_numbers(d6, 'E_VALORMATRICULAUNIVERSIDAD', mapping_matri)

print(d6['E_VALORMATRICULAUNIVERSIDAD'].unique())
print(d6['E_VALORMATRICULAUNIVERSIDAD'].isna().sum(), "valores quedaron como NaN")
```

```python
import re
d7 = d6.copy()

def clean_text_column(df, column):
    df[column] = (
        df[column]
        .astype(str)
        .str.lower()
        .str.normalize('NFKD')
        .str.encode('ascii', errors='ignore')
        .str.decode('utf-8')
        .str.replace(r'[^a-z0-9\s]', '', regex=True)
        .str.replace(r'\s+', '', regex=True)
    )
    return df

mapping_work = {
    '0': 1,
    'menosde10horas': 2,
    'entre11y20horas': 3,
    'entre21y30horas': 4,
    'masde30horas': 5,
}

d7 = clean_text_column(d7, 'E_HORASSEMANATRABAJA')
d7 = replace_category_for_numbers(d7, 'E_HORASSEMANATRABAJA', mapping_work)
```
